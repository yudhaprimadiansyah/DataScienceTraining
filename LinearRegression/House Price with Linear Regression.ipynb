```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import math
```

```python
data = pd.read_csv("house_prices.csv")
x = np.array(data["sqft_living"]).reshape(-1, 1)
y = np.array(data["price"]).reshape(-1,1)

model = LinearRegression()
model.fit(x,y)
print(math.sqrt(mean_squared_error(x,y)))
print(model.score(x,y))
```

```python
plt.scatter(x,y,color="green")
plt.plot(x, model.predict(x), color="red")
plt.title("Harga Rumah Berdasarkan SQFT Living")
plt.xlabel("SQFT Living")
plt.ylabel("Harga")
plt.show()
```

```python
print("Coeficient : ", model.coef_[0])
print("Intercept : ", model.intercept_[0])
```

```python
print("Prediksi harga rumah dengan SQFT 50000 : ", model.predict([[50000]]))
```
