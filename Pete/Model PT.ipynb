```
#jalankan script ini dahulu untuk memulai apapun
import keras.models
import os
import pandas as pd
```

```
# Mari Templating CSV untuk klasifikasinya
sDir = 'DataSets'
dirs = ['a01']#os.listdir(sDir)
csvCand = "Features,Target\n"
for i in dirs:
    treeDirs = sDir+'/'+str(i)
    sTreeDirs = os.listdir(treeDirs)
    for c in sTreeDirs:
      #print(str(i)+'/'+str(c))
      sTargetDir = 'DataSets/a01' #!!
      targetDir = os.listdir(sTargetDir)
      for d in targetDir:
        #print(sTargetDir+'/'+str(d))
        sImagePath = sTargetDir+'/'+str(d)
        impath = os.listdir(sImagePath)
        for x in impath:
            image = sImagePath+'/'+x
            #print(image)
            csvCand += (image+','+str(d)+'\n')

csvv = open('model_data.csv', 'w')
csvv.write(csvCand)
csvv.close()
```
